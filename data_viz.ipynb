```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('melb_data.csv')
df.head()
```

```python
sns.histplot(data=df, x='Price')
plt.title('Distribusi Harga Rumah', size=15)
plt.grid()
```

```python
sns.boxplot(data=df, x='Distance', y='Regionname', orient ='h')
plt.title('Distribusi Jarak le CBD berdasarkan Nama Region', size=15)
```

```python
sns.barplot(data=df, x='Type', y='Price', estimator = np.median)
plt.title('Perbandingan Nilai median Harga Rumah', size=15)
```

```python
sns.scatterplot(data=df, x='Price', y='Distance', hue='Type')
plt.title('Korelasi Harga Rumah dan Jarak Rumah ke CBD', size=15)
```
